```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results_final_2.txt')
df_mapping = pd.DataFrame({'MODELNAME': ['TRPO_BAYRN', 'TRPO_UDR', 'TRPO', 'PPO', 'ACTOR_CRITIC', 'REINFORCE'],})
sort_mapping = df_mapping.reset_index().set_index('MODELNAME')
df['modelname_mapping'] = df['MODELNAME'].map(sort_mapping['index'])
df = df.sort_values('modelname_mapping')
```

```python
plot, ax = plt.subplots(figsize = (11,6))
f = sns.violinplot(x='MODELNAME', y='REWARD', data=df, ax=ax)
f.set_xticklabels(f.get_xticklabels(), fontsize=14)
f.set_xlabel("")
f.set_yticklabels(f.get_yticks(), fontsize=14)
f.set_ylabel(f.get_ylabel(), fontsize=14)
```
