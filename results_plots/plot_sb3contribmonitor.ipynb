```python
from stable_baselines3.common.results_plotter import load_results, ts2xy, plot_results
import seaborn as sns
from stable_baselines3.common import results_plotter
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
aviablelist = [[model.split('/')[1].split('-')[0],model] for model in glob.glob('logs/*')]
aviablelist
```

```python
log_folder = "logs/"
ppologs = load_results([x[1] for x in aviablelist if x[0]=='PPO'][0])
trpologs = load_results([x[1] for x in aviablelist if x[0]=='TRPO'][0])
```

```python
from matplotlib.pyplot import legend


fig, ax = plt.subplots(1,1,figsize=(13,8))
sns.regplot(data = ppologs, x='index', y='r', ax=ax, ci=90, color='orange', label='PPO', scatter_kws={'alpha':0.2}, line_kws={'linewidth':5})
f = sns.regplot(data = trpologs, x='index', y='r', ax=ax, ci=90, label='TRPO', scatter_kws={'alpha':0.2},line_kws={'linewidth':5})
ax.legend(loc='best', fontsize=16)
f.set_xticklabels(f.get_xticklabels(), fontsize=12)
f.set_xlabel('time', fontsize=16)
f.set_yticklabels(f.get_yticks(), fontsize=12)
f.set_ylabel('return', fontsize=16)
#ax.set_title("Return over time", fontsize=16)
```
